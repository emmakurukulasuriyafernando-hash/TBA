```python
import math

# Define constants for the spiral simulation
MAX_RADIUS = 100.0  # Maximum radius of the spiral
MIN_RADIUS = 10.0   # Minimum radius of the spiral
NUM_ITERATIONS = 200 # Number of steps for each spiral phase
DEGREES_PER_ITERATION = 5 # How many degrees the robot turns each iteration

print(f"Constants defined for spiral simulation: MAX_RADIUS={MAX_RADIUS}, MIN_RADIUS={MIN_RADIUS}, NUM_ITERATIONS={NUM_ITERATIONS}, DEGREES_PER_ITERATION={DEGREES_PER_ITERATION}")

# Initialize lists to store trajectory data
inward_radii = []
inward_angles = []
outward_radii = []
outward_angles = []

# --- Inward Spiral (Outside to Center) ---
print("\n--- Simulating Inward Spiral Movement ---")

current_radius = MAX_RADIUS
current_angle = 0.0 # Start at 0 degrees

for i in range(NUM_ITERATIONS):
    current_radius = MAX_RADIUS - (i / NUM_ITERATIONS) * (MAX_RADIUS - MIN_RADIUS)
    current_angle += DEGREES_PER_ITERATION

    motor_speed = int(min(255, current_radius * 2.5))
    set_motor_speed(motor_speed)
    set_motor_direction(int(current_angle))

    inward_radii.append(current_radius)
    inward_angles.append(current_angle)

    # print(f"Iteration {i+1}: Radius={current_radius:.2f}, Angle={current_angle:.2f} degrees")

# --- Outward Spiral (Center to Outside) ---
print("\n--- Simulating Outward Spiral Movement ---")

# Continue from the final state of the inward spiral
# current_radius is already set to the end of inward spiral, but let's make sure it's MIN_RADIUS
current_radius_outward = MIN_RADIUS
current_angle_outward = current_angle # Continue angle from inward spiral's end

for i in range(NUM_ITERATIONS):
    current_radius_outward = MIN_RADIUS + (i / NUM_ITERATIONS) * (MAX_RADIUS - MIN_RADIUS)
    current_angle_outward += DEGREES_PER_ITERATION

    motor_speed = int(min(255, current_radius_outward * 2.5))
    set_motor_speed(motor_speed)
    set_motor_direction(int(current_angle_outward))

    outward_radii.append(current_radius_outward)
    outward_angles.append(current_angle_outward)

    # print(f"Iteration {i+1}: Radius={current_radius_outward:.2f}, Angle={current_angle_outward:.2f} degrees")

print("\n--- Trajectory Data Collection Summary ---")
print(f"Inward Radii (first 5): {inward_radii[:5]}")
print(f"Inward Angles (first 5): {inward_angles[:5]}")
print(f"Length of Inward Radii: {len(inward_radii)}")
print(f"Length of Inward Angles: {len(inward_angles)}")

print(f"\nOutward Radii (first 5): {outward_radii[:5]}")
print(f"Outward Angles (first 5): {outward_angles[:5]}")
print(f"Length of Outward Radii: {len(outward_radii)}")
print(f"Length of Outward Angles: {len(outward_angles)}")
```
